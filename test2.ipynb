```python
import numpy as np

#훈련 데이터 셋을 랜덤으로 생성합니다.
#shape = uid, iid, rate
train = np.random.randint(0,2, size=(500,3))
train.shape
```

```python
import pandas as pd
train = pd.DataFrame(train)
```

```python
#column name변경
train = train.rename(columns={0:'uid', 1:'iid', 2:'rate'})
```

```python
train['uid'] = np.random.randint(0, 200, size= (500,1))
train['iid'] = np.random.randint(0, 200, size= (500,1))
```

```python
train['rate'] = np.ones((500,1))
```

```python
train
```

```python
from surprise import Dataset
from surprise import Reader
from surprise.model_selection import train_test_split
from surprise import accuracy
reader = Reader(rating_scale = (0,1))
data = Dataset.load_from_df(train, reader = reader)
x_train, x_test = train_test_split(data, test_size = 0.25, random_state= 42)
```

```python
from surprise import SVD
algo = SVD()
algo.fit(x_train)
```

```python
algo.test(x_test)
```

```python
accuracy.rmse(algo.test(x_test))
```

```python
algo.predict(1,100)
```

```python
for i in range(1,200):
    uid, iid, r_ui, est, details = algo.predict(1, i)
    if est < 0.8:
        print(uid, iid, est)
```

```python
for i in range(1,200):
    uid, iid, r_ui, est, details = algo.predict(0, i)
    print(f'{uid}번 유저가 {iid}번 유저를 찜할 확률은 {est}입니다')
```
